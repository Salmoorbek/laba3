```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_function(x):
    return 2*x**2 + 3*x + 1

def harmonic_function(x):
    return np.sin(x)

t = np.linspace(0, 4*np.pi, 1000)
x_parametric = np.cos(3*t)
y_parametric = np.sin(2*t)

years = list(range(2012, 2023))
min_expenses = [4341.1, 4599.2, 4981.5, 5182.9, 4794.3, 4900.7, 4792.5, 4806.3, 5358.5, 6268.3, 7178.3]

plt.figure(figsize=(15, 12))

plt.subplot(3, 3, 1)
x_quad = np.linspace(-10, 10, 100)
plt.plot(x_quad, quadratic_function(x_quad))
plt.title('  ')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(3, 3, 2)
x_harm = np.linspace(0, 2*np.pi, 100)
plt.plot(x_harm, harmonic_function(x_harm))
plt.title('  ')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(3, 3, 3)
plt.plot(x_quad, quadratic_function(x_quad), label=' ')
plt.title('   ')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.subplot(3, 3, 4)
plt.plot(x_harm, harmonic_function(x_harm), label=' ')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.subplot(3, 3, 5)
plt.plot(x_quad, quadratic_function(x_quad), label=' ')
plt.plot(x_harm, harmonic_function(x_harm), label=' ')
plt.title('    ')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.subplot(3, 3, 6)
plt.plot(x_parametric, y_parametric, label=' ')
plt.title('  ')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.subplot(3, 3, 7)
plt.bar(years, min_expenses, color='skyblue')
plt.title(' ')
plt.xlabel('')
plt.ylabel(' ')

plt.subplot(3, 3, 8)
plt.pie(min_expenses, labels=years, autopct='%1.1f%%')
plt.title(' ')

ax = plt.subplot(3, 3, 9, projection='3d')
X, Y = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))
Z = np.sin(np.sqrt(X**2 + Y**2))
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title(' ')

plt.tight_layout()
plt.show()
```
